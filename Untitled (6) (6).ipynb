```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor


# Define models
models = { 'Neural Network Regression': MLPRegressor(max_iter=1000)}

# Initialize DataFrame for storing results
results = pd.DataFrame(columns=['Model', 'RMSE Train', 'MAE Train', 'R2 Train', 'RMSE Test', 'MAE Test', 'R2 Test'])

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)

    # Metrics
    rmse_train = np.sqrt(mean_squared_error(y_train, pred_train))
    mae_train = mean_absolute_error(y_train, pred_train)
    r2_train = r2_score(y_train, pred_train)

    rmse_test = np.sqrt(mean_squared_error(y_test, pred_test))
    mae_test = mean_absolute_error(y_test, pred_test)
    r2_test = r2_score(y_test, pred_test)

    # Create a temporary DataFrame with the results
    temp_df = pd.DataFrame([{
        'Model': name,
        'RMSE Train': rmse_train,
        'MAE Train': mae_train,
        'R2 Train': r2_train,
        'RMSE Test': rmse_test,
        'MAE Test': mae_test,
        'R2 Test': r2_test
    }])

    # Append the temporary DataFrame to results
    results = pd.concat([results, temp_df], ignore_index=True)

# Sort results by RMSE Test
results_sorted_rmse_test = results.sort_values(by='RMSE Test')
results_sorted_rmse_train = results.sort_values(by='RMSE Train')
results_sorted_r2_test = results.sort_values(by='R2 Test')

# Display results
print("Model Performance Sorted by RMSE Test:")
print(results_sorted_rmse_test)

print("\nModel Performance Sorted by RMSE Train:")
print(results_sorted_rmse_train)

print("\nModel Performance Sorted by R2 Test:")
print(results_sorted_r2_test)

# Plot training vs testing metrics
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

# Plot RMSE Train
sns.barplot(x='Model', y='RMSE Train', data=results_sorted_rmse_train, ax=ax[0])
ax[0].set_title('RMSE Train (Ascending Order)')
ax[0].tick_params(axis='x', rotation=90)

# Plot RMSE Test
sns.barplot(x='Model', y='RMSE Test', data=results_sorted_rmse_test, ax=ax[1])
ax[1].set_title('RMSE Test (Ascending Order)')
ax[1].tick_params(axis='x', rotation=90)

# Plot R2 Test
sns.barplot(x='Model', y='R2 Test', data=results_sorted_r2_test, ax=ax[2])
ax[2].set_title('R2 Test (Ascending Order)')
ax[2].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()

# Display actual vs predicted values for the best model
best_model_name = results_sorted_rmse_test.iloc[0]['Model']
best_model = models[best_model_name]
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

# Create DataFrame for actual vs predicted values
comparison_df = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})
print("\nActual vs Predicted values for the best model ({}):".format(best_model_name))
print(comparison_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor

# Define models
models = {
    'Neural Network Regression': MLPRegressor(max_iter=1000)
}

# Initialize DataFrame for storing results
results = pd.DataFrame(columns=['Model', 'RMSE Train', 'MAE Train', 'R2 Train', 'RMSE Test', 'MAE Test', 'R2 Test'])

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)

    # Metrics
    rmse_train = np.sqrt(mean_squared_error(y_train, pred_train))
    mae_train = mean_absolute_error(y_train, pred_train)
    r2_train = r2_score(y_train, pred_train)

    rmse_test = np.sqrt(mean_squared_error(y_test, pred_test))
    mae_test = mean_absolute_error(y_test, pred_test)
    r2_test = r2_score(y_test, pred_test)

    # Create a temporary DataFrame with the results
    temp_df = pd.DataFrame([{
        'Model': name,
        'RMSE Train': rmse_train,
        'MAE Train': mae_train,
        'R2 Train': r2_train,
        'RMSE Test': rmse_test,
        'MAE Test': mae_test,
        'R2 Test': r2_test
    }])

    # Append the temporary DataFrame to results
    results = pd.concat([results, temp_df], ignore_index=True)

# Plot metrics versus models
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

# RMSE Train vs Models
sns.lineplot(x='Model', y='RMSE Train', data=results, marker='o', ax=ax[0], label='RMSE Train', color='blue')
sns.lineplot(x='Model', y='RMSE Test', data=results, marker='o', ax=ax[0], label='RMSE Test', color='red')
ax[0].set_title('RMSE vs Models')
ax[0].tick_params(axis='x', rotation=90)
ax[0].legend()

# MAE Train vs Models
sns.lineplot(x='Model', y='MAE Train', data=results, marker='o', ax=ax[1], label='MAE Train', color='blue')
sns.lineplot(x='Model', y='MAE Test', data=results, marker='o', ax=ax[1], label='MAE Test', color='red')
ax[1].set_title('MAE vs Models')
ax[1].tick_params(axis='x', rotation=90)
ax[1].legend()

# R2 Test vs Models
sns.lineplot(x='Model', y='R2 Train', data=results, marker='o', ax=ax[2], label='R2 Train', color='blue')
sns.lineplot(x='Model', y='R2 Test', data=results, marker='o', ax=ax[2], label='R2 Test', color='red')
ax[2].set_title('R2 vs Models')
ax[2].tick_params(axis='x', rotation=90)
ax[2].legend()

plt.tight_layout()
plt.show()

# Display actual vs predicted values for the best model
best_model_name = results.loc[results['RMSE Test'].idxmin(), 'Model']
best_model = models[best_model_name]
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

# Create DataFrame for actual vs predicted values
comparison_df = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})
print("\nActual vs Predicted values for the best model ({}):".format(best_model_name))
print(comparison_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor




# Define models
models = {
    'Neural Network Regression': MLPRegressor(max_iter=1000), 
}

# Initialize DataFrame for storing results
results = pd.DataFrame(columns=['Model', 'RMSE Train', 'MAE Train', 'R2 Train', 'RMSE Test', 'MAE Test', 'R2 Test'])

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)

    # Metrics
    rmse_train = np.sqrt(mean_squared_error(y_train, pred_train))
    mae_train = mean_absolute_error(y_train, pred_train)
    r2_train = r2_score(y_train, pred_train)

    rmse_test = np.sqrt(mean_squared_error(y_test, pred_test))
    mae_test = mean_absolute_error(y_test, pred_test)
    r2_test = r2_score(y_test, pred_test)

    # Create a temporary DataFrame with the results
    temp_df = pd.DataFrame([{
        'Model': name,
        'RMSE Train': rmse_train,
        'MAE Train': mae_train,
        'R2 Train': r2_train,
        'RMSE Test': rmse_test,
        'MAE Test': mae_test,
        'R2 Test': r2_test
    }])

    # Append the temporary DataFrame to results
    results = pd.concat([results, temp_df], ignore_index=True)

# Sort results by RMSE Test
results_sorted = results.sort_values(by='RMSE Test')

# Display results
print("Results sorted by RMSE Test:")
print(results_sorted)

# Bar plots for metrics
fig, axes = plt.subplots(3, 1, figsize=(10, 15), sharex=True)
metrics = ['RMSE Test', 'MAE Test', 'R2 Test']
colors = ['blue', 'green', 'red']

for ax, metric, color in zip(axes, metrics, colors):
    sns.barplot(x='Model', y=metric, data=results_sorted, ax=ax, palette=[color])
    ax.set_title(f'{metric} vs Models')
    ax.set_xlabel('Model')
    ax.set_ylabel(metric)
    ax.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()

# Scatter plots for predicted vs actual values
for name, model in models.items():
    model.fit(X_train, y_train)
    pred_test = model.predict(X_test)

    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, pred_test, alpha=0.7)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', color='red')
    plt.title(f'{name}: Predicted vs Actual')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.grid(True)
    plt.show()
```
